## Sorting, Ranking & Counting in Pandas

```python
import pandas as pd
```

```python
data = {
    "name": ["Alice", "Bob", "Caro", "Dan"],
    "age": [23, None, 22, 24],
    "score": [88, 72, None, 91]
}

df = pd.DataFrame(data)
df
```

## Sorting

### Sort by Values

```python
df.sort_values("score") # sort dataframe by score in ascending order
```

```python
df.sort_values("score", ascending=False) # sort dataframe by score in descending order
```

```python
# sort multiple columns at once
df.sort_values(["score", "age"])
```

```python
# sort df by score in ascending order, then age in descending order
df.sort_values(["score", "age"], ascending=[True, False])
```

### Sort by Index

```python
df.sort_index() # also accepts an ascending argument
```

Sort_index is useful after filtering or group operations.

## Ranking

When ranking a dataframe column in ascending order, the smallest value gets ranked first, the second smallest ranked second, and so on. In descending order, the largest value gets ranked first, the second largest gets ranked second, and so on.

```python
df["rank"] = df["score"].rank() # default ascending order
df
```

```python
df["rank"] = df["score"].rank(ascending=False) # descending order
df
```

### Ranking method 1: Average (Default)

In case of a tie, the two ranks occupied by the tied scores are replaced by their mean. For example, the two 80s occupy ranks 2 and 3, so Pandas assigns them a shared rank of (2 + 3) / 2 = 2.5

```python
r_scores = pd.DataFrame({"scores": [90, 80, 80, 70]})
r_scores["rank"] = r_scores["scores"].rank(ascending=False, method="average") # rank scores in descending order
r_scores
```

### Ranking Method 2: Min

In case of a tie, the two ranks occupied by the tied scores are replaced by their minimum. For example, the two 80s occupy ranks 2 and 3, so Pandas assigns them a shared rank of min(2, 3) = 2

```python
r_scores["rank"] = r_scores["scores"].rank(ascending=False, method="min")
r_scores
```

### Ranking Method 3: Max

In case of a tie, the two ranks occupied by the tied scores are replaced by their maximum. For example, the two 80s occupy ranks 2 and 3, so Pandas assigns them a shared rank of max(2, 3) = 3

```python
r_scores["rank"] = r_scores["scores"].rank(ascending=False, method="max")
r_scores
```

### Ranking Method 4: Dense

In case of a tie, no ranks are skipped afterward. For example, the two 80s occupy a shared rank of 2, and the succeeding value is assigned a rank of 3. There are no gaps in-between like in the other methods.

```python
r_scores["rank"] = r_scores["scores"].rank(ascending=False, method="dense")
r_scores
```

## Binning

Binning is a Pandas function used to convert continuous numeric variables into categories. It is an example of feature engineering.

Binning lets you:

- summarize distributions

- compare groups

- simplify analysis

- prepare data for ML models

```python
exam_scores = pd.Series([45, 67, 82, 91])
pd.cut(exam_scores, bins=[0, 60, 80, 100])
```

Bins created:
- (0, 60]

- (60, 80]

- (80, 100]

pd.cut is left-exclusive, right-inclusive by default, but this behaviour can be changed by specifying the argument right=False

### Adding Labels

```python
pd.cut(
    exam_scores,
    bins=[0, 60, 80, 100],
    labels=["Low", "Medium", "High"]
)
```

### Automatic Binning

```python
pd.cut(exam_scores, bins=3, labels=["below average", "average", "above average"])
```

### Percentile-based Binning

```python
pd.qcut(exam_scores, q=3)
```
